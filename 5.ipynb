```python
import numpy as np
import pandas as pd
```

```python
# Read in txt file
stacks_list = [ [] for _ in range(9)] 
stacks_positions = np.arange(0, 37, 4)
with open('input_2.txt') as f:
    for line in f:
        for stack in range(0, 9):
            current_box = line[stacks_positions[stack]:stacks_positions[stack]+3].strip()
            if ('[' or ']') in current_box:
                stacks_list[stack].insert(0, current_box)
                
```

```python
with open('input.txt') as f:
    commands = f.readlines()
```

```python
for command in commands:
    command = command.strip().split(' ')
    print(command)
    how_many = int(command[1])
    from_where = int(command[3])
    to_where = int(command[5])
    grabbed_stack = []
    for _ in range(how_many):
        grabbed_stack.insert(0, stacks_list[from_where-1].pop())
    stacks_list[to_where-1].extend(grabbed_stack)
```

```python
result = ''
for stack in stacks_list:
    result += ''.join(stack[-1][1])
result
```
