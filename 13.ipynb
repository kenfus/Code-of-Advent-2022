```python
def compare_lists(first_list, second_list):
    print(first_list, second_list)
    if type(first_list) == int:
        first_list = [first_list]
    if type(second_list) == int:
        second_list = [second_list]
        
    # Compare lists
    current_state = True
    for i, left_side in enumerate(first_list):
        try:
            right_side = second_list[i]
        except IndexError:
            print('Right side run out of elements')
            return False
        if isinstance(left_side, int) and isinstance(right_side, int):
            if left_side < right_side:
                print('Left side is smaller')
                return True
            if left_side > right_side:
                print('Right side is smaller')
                return False
            if left_side == right_side:
                print('Elements are equal')
                continue
        else:
            print('Recursion')
            current_state  = compare_lists(left_side, right_side)
    # Left side run out of elements
    print('Left side run out of elements')
    return current_state
            
```

```python
compare_lists([1,[2,[3,[4,[5,6,7]]]],8,9], [1,[2,[3,[4,[5,6,0]]]],8,9])
```

```python
first_list = None
second_list = None
with open('input.txt') as f:
    data = f.readlines()
    
right_order = []
pair = 1
for line in data:
    if first_list is None:
        first_list = eval(line.strip())
        continue
    elif second_list is None:
        second_list = eval(line.strip())
        continue
    else:
        result = compare_lists(first_list, second_list)
        print(f'Pair {pair}: {result}')
        if result:
            right_order.append(pair)
        pair += 1
        first_list = None
        second_list = None
```
